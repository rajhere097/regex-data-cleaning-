```python
import re
# library for regular expression 
```

```python
# Exercise 1: Matching a date in the format "MM/DD/YYYY"

text1 = "Today's date is 03/29/2024"
```

```python
#Ans
pattern1 = r"\d{2}/\d{2}/\d{4}"
print("Exercise 1:", re.search(pattern1, text1).group())
```

```python
# My answer
result = re.search(r'\d{2}/\d{2}/\d{4}', text1)
if result:
    print(result.group())
```

```python
# Exercise 2: Matching a phone number in the format "XXX-XXX-XXXX"

text2 = "Contact me at 123-456-7890"
```

```python
#  Ans
pattern2 = r"\d{3}-\d{3}-\d{4}"
print("Exercise 2:", re.search(pattern2, text2).group())
```

```python
# Exercise 3: Matching email addresses

text3 = "My email is example@email.com"
```

```python
#Ans
pattern3 = r"\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b"

print("Exercise 3:", re.findall(pattern3, text3))

# pattern can also be r'\b\w+@\w+.\w{3}'
```

```python
# Exercise 4: Matching hexadecimal color codes

text4 = "The color of the sky is #00BFFF"
```

```python
#Ans
pattern4 = r"#[0-9a-fA-F]{6}"
print("Exercise 4:", re.findall(pattern4, text4))
```

```python
# Exercise 5: Matching a URL

text5 = "Visit our website at https://example.com"
```

```python
#Ans
pattern5 = r"https?://(?:[-\w.]|(?:%[\da-fA-F]{2}))+"
print("Exercise 5:", re.search(pattern5, text5).group())

# this can also work: r"https?://(?:\w+.\w{3})"
```

```python
# Exercise 6: Matching words containing "cat"

text6 = "The cat sat on the mat"
```

```python
#Ans
pattern6 = r"\b\w*cat\w*\b"
print("Exercise 6:", re.findall(pattern6, text6))
```

```python
# Exercise 7: Matching 3-letter words

text7 = "The fox ran fast"
```

```python
#Ans
pattern7 = r"\b\w{3}\b"
print("Exercise 7:", re.findall(pattern7, text7))
```

```python
# Exercise 8: Matching a repeating word
text8 = "This is is a repeated word"
```

```python
#Ans
pattern8 = r"\b(\w+)\s+\1\b"
print("Exercise 8:", re.search(pattern8, text8).group())
```

```python
# Exercise 9: Matching sentences ending with a punctuation

text9 = "This is a sentence. And another one!"
```

```python
#Ans
pattern9 = r"[^.!?]+[.!?]"
print("Exercise 9:", re.findall(pattern9, text9))
```

```python
# Exercise 10: Matching HTML tags

text10 = "<h1>Hello</h1><p>This is a paragraph</p>"
```

```python
#ans
pattern10 = r"<(\w+)>.*?</\1>"
print("Exercise 10:", re.findall(pattern10, text10))
```

```python

# Exercise 11: Matching lines with only digits

text11 = "123\n45\n6\n789"
```

```python
#ans
pattern11 = r"^\d+$"
print("Exercise 11:", re.findall(pattern11, text11, re.MULTILINE))
```

```python
# Exercise 12: Matching strings containing "abc" or "def"

text12 = "The alphabet contains abc and def"
```

```python
#Ans
pattern12 = r"abc|def"
print("Exercise 12:", re.findall(pattern12, text12))
```

```python
# Exercise 13: Matching words starting with a vowel

text13 = "An apple a day keeps the doctor away"

```

```python
#Ans
pattern13 = r"\b[aeiouAEIOU]\w*\b"
print("Exercise 13:", re.findall(pattern13, text13))
```

```python

# Exercise 14: Matching lines containing only letters and spaces

text14 = "This is a line\n123\nThis is another line"
```

```python
#Ans
pattern14 = r"^[a-zA-Z\s]+$"
print("Exercise 14:", re.findall(pattern14, text14, re.MULTILINE))
```

```python
# Exercise 15: Matching words with at least two consecutive vowels

text15 = "The owl is sleeping peacefully"
```

```python

#Ans
pattern15 = r"\b\w*[aeiou]{2,}\w*\b"
print("Exercise 15:", re.findall(pattern15, text15))
```

```python
# Exercise 17: Matching time in the format "HH:MM AM/PM"

text17 = "The meeting starts at 10:30 AM"

```
